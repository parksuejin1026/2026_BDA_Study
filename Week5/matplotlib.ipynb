```python
from typing import ChainMap
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2] # or 논리연산자
y = y[mask1 + mask2]

# x와 y의 절댓값이 50이 넘을 때만 출력하도록 함
plt.scatter(x, y, c = x, cmap = 'jet', alpha = 0.4)
plt.colorbar()
plt.show()
```

```python
from typing import ChainMap
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 * mask2] # and 논리연산자
y = y[mask1 * mask2]

# x와 y의 절댓값이 50이 넘을 때만 출력하도록 함
plt.scatter(x, y, c = x, cmap = 'jet', alpha = 0.5)
plt.colorbar()
plt.show()
```

# fig의 크기를 지정하는 방법

```python
# plt.subplots에서 figure 크기 지정
fig, axes = plt.subplots(2, 1, figsize = (8, 6))


# 각 서브플롯에 데이터 추가
axes[0].plot([1, 2, 3,], [1, 4, 9])
axes[1].plot([1, 2, 3], [1, 2, 3])

plt.show()
```

```python
# plt.subplots()에서 반환 받은 fig에서 크기 지정

# plt.subplots()로 figure와 크기를 지정
fig, axes = plt.subplots(2, 1, figsize = (8, 6))
fig.set_size_inches(10,4)

# 각 서브플롯에 데이터 추가
axes[0].plot([1, 2, 3], [1, 4, 9])
axes[1].plot([1, 2, 3], [1, 2, 3])
```

```python
fig = plt.figure()

fig.add_subplot(321).set_title('321') # 3행 2열의 1번째 인덱스 그래프의 제목을 321로 한다
fig.add_subplot(322).set_title('322')
fig.add_subplot(323).set_title('323')
fig.add_subplot(324).set_title('324')
fig.add_subplot(325).set_title('325')
fig.add_subplot(326).set_title('326')

fig.subplots_adjust(wspace = .2, hspace= .9) # 수평 길이 .2 수직 길이 .9
plt.show()
```

```python
plt.suptitle('전체 제목', fontsize = 14, y = 1.02)

plt.subplot(2, 3 , 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱수')

plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4])
plt.title('선형')

plt.subplot(2, 3 , 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize = (7, 4))
fig.set_size_inches(6, 3)

ax[0, 1].plot([1, 4, 9, 16])
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])

plt.suptitle('전체 제목' , fontsize = 14, fontweight = 'bold')

ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(6, 4))
ax1 = fig.add_subplot(231)
ax1.plot([1, 4, 9, 16])
ax1.set_title('제곱수')
ax2 = fig.add_subplot(233)
ax2.plot([1, 2, 3, 4])
ax2.set_title('선형')
ax3 = fig.add_subplot(235)
ax3.plot([6, 5, 2, 1])
ax3.set_title('감소')
fig.suptitle('전체 제목', fontsize=14)
plt.tight_layout()
plt.show()
```
