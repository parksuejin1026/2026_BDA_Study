```python
basket = {'apple', 'orange', 'apple', 'pear'} # 집합은 중복이 불가능
print(basket)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}

print(A.union(B)) # 합집합
print(A.intersection(B)) # 교집합
print(A.difference(B)) # 차집합
print(B.difference(A)) # 차집합
print(A.symmetric_difference(B)) # 여집합
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6, 7}

print(A | B) # A, B 집합의 모든 값, 합집합
print(A & B) # A, B 집합의 요소에서 중복되 값만, 교집합
print(A - B) # A 집합을 기준으로 B 집합에서의 중복되는 값을 제외한 값, 차집합
print(B - A) # B 집합을 기준으로 A 집합에서의 중복되는 값을 제외한 값, 차집합
print(A ^ B) # A, B 집합의 요소에서 중복되는 값을 제외한 값만, 여집합
```
