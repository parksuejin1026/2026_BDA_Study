```python
rg.random((4, 3)) # 0 이상 1 미만의 실수 난수로 구성된 4행 3열 배열 생성
```

```python
rg = np.random.default_rng(12345) # 난수 생성기를 초기화하고 seed 값 설정
r = rg.integers(low = 0, high = 10, size = 3) # low(포함)부터 high(미포함) 사이의 정수 난수 3개 생성
r
```

```python
rg.integers(1, 7, (3,4)) # 1(포함)부터 7(미포함) 사이의 정수 난수로 구성된 3행 4열 배열 생성
```

```python
rg = np.random.default_rng(seed = 42) # 난수 생성기를 초기화하고 seed 값 설정
rg.standard_normal(10) # 표준 정규 분포(평균 0, 표준편차 1)를 따르는 실수 난수 10개 생성
```

```python
rg.standard_normal((4, 3)) # 표준 정규 분포를 따르는 실수 난수로 구성된 4행 3열 배열 생성
```

```python
rg.standard_normal((10, 6)) # 표준 정규 분포를 따르는 실수 난수로 구성된 10행 6열 배열 생성
```

```python
x = rg.standard_normal((10, 6))
for i in x:
 # 배열 i에서 2.0 이상인 요소가 하나라도 있는지 확인
 if (i >= 2.0).any():
  print(i)
```
